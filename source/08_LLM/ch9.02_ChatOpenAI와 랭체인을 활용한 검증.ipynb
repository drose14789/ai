## 2.1 OpenAI SDK를 선택
- pip install openai

```python
from openai import OpenAI
client = OpenAI()
response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages=[
        {"role":"system", "content": "너는 친절하게 대답해 주는 비서야"},
        {"role":"user", "content":"2020년 월드 시리즈 누가 우승했어?"}
    ],
    temperature=0.5, #0~0.3:일관되고 예측가능한답 / 0.8~1:창의적다양한단어답볍ㄴ
)
```

```python
response.choices[0].message.content
```

```python
response.usage # 사용 토큰수
```

## 2.2 LangChain을 선택

```python
from langchain_openai import ChatOpenAI
# .env에 OPENAI_API_KEY 등록
llm = ChatOpenAI(
    model = "gpt-4o-mini",
    temperature=0.5
)
```

```python
ai_message = llm.invoke("2020년 월드 시리즈는 누가 우승했어요?")
```

```python
ai_message.content
```

```python
ai_message.usage_metadata
```

```python
from langchain_core.messages import SystemMessage, HumanMessage
messages = [
     SystemMessage(content="너는 친절하게 대답해주는 비서야"),
     HumanMessage(content="2020년 월드 시리즈 누가 우승했어?")
]
```

```python
ai_message = llm.invoke(messages)
print(ai_message.content)
```
