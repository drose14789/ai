```python
# 복합체인 : 나라명 -> 그 나라에서 제일 유명한 음식(fodd_chain)
#                        음식 -> 레시피(recipe_chain)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
llm = ChatOllama(model="llama3.2:1b",
                temperature=0) # 일관된 답변(보수적인 답변)
output_parser = StrOutputParser() # AIMessage()를 Str변환
```

```python
food_prompt = PromptTemplate(
    template = """Please recommend the most famous food of the country based on the following informat:
    {country}
    Return the name of the food only
    """,
    input_variables=["country"]
)
country=input("나라")
output_parser.invoke(llm.invoke(food_prompt.invoke({"country":country})))
```

```python
food_chain = food_prompt | llm | output_parser
```

```python
recipe_prompt = PromptTemplate(
    template = """Please let me know the recipe based on the following informat:
    {food}
    """,
    input_variables=["food"]
)
output_parser.invoke(llm.invoke(recipe_prompt.invoke({"food":food})))
```

```python
recipe_chain = recipe_prompt | llm | output_parser
```

```python
final_chain = food_chain | recipe_chain
final_chain.invoke({"country":country, "food":food})
```

```python
from langchain_core.runnables import RunnablePassthrough
final_chain = (
    {"food": food_chain, "country": RunnablePassthrough()} 
    | recipe_chain
)
```

```python
result =final_chain.invoke({"country": country})
```
