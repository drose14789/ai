# 2. LLM 답변 생성
- ollama.com에서 모델 선택(llama / exaone)
    * powershell이나 cmd창에서 ollama pull llama3.2:1b
    * powershell이나 cmd창에서 ollama pull exaone3.5:2.4b

```python
from langchain_ollama import ChatOllama
from langchain_core.messages import SystemMessage, HumanMessage
llm = ChatOllama(
    model="llama3.2:1b",
)
messages = [
    SystemMessage(content= "너는 친절하게 대답해 주는 비서야"),
    HumanMessage(content = "2020년 월드 시리즈 누가 우승했어?")
]
```

```python
ai_message = llm.invoke(input=messages)
ai_message
```
